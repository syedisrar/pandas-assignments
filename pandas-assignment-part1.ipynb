```python
import numpy as np
import pandas as pd
import glob
```

```python
census_files=glob.glob("./states*.csv")
allcsv = []
for csv_file in census_files:
    print(csv_file)
    df = pd.read_csv(csv_file, index_col=None, header=0)
    allcsv.append(df)
us_census = pd.concat(allcsv, axis=0, ignore_index=True)
us_census.head()
    
```

```python
us_census.columns
```

```python
us_census.dtypes
```

```python
us_census.drop(columns = ['Unnamed: 0'], inplace=True)
```

```python
us_census.head()
```

```python
us_census[["Male","Female"]] =  us_census.GenderPop.str.split("_", expand=True)
```

```python
us_census.head(5)
```

```python
us_census.drop(columns = ["GenderPop"], inplace = True)
```

```python
us_census.head(5)
```

```python
us_census["Income"] = us_census.Income.str.strip("$")
```

```python
us_census.head(1)
```

```python
us_census = us_census.replace ("%", "", regex=True)
us_census.head(1)
```

```python
us_census = us_census.replace ("M", "", regex=True)
us_census = us_census.replace ("F", "", regex=True)
us_census.head(1)
```

```python
us_census.loc[:, 'Hispanic':'Income'] = round (us_census.loc [:, 'Hispanic' : 'Income'].apply(pd.to_numeric),2)
us_census.head()
```

```python
us_census.dtypes
```

```python
us_census["Male"] = us_census["Male"].astype(int)
us_census.head()
```

```python
us_census.drop(columns = ['Female'], inplace=True)
us_census['Female'] = us_census['TotalPop'] - us_census['Male']
us_census.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter (us_census["Female"], us_census["Income"])
plt.xlabel ('Salary $', fontsize = 16)
plt.ylabel ('Female Population', fontsize = 16)
plt.show()
```

```python
us_census.duplicated().any()
```

```python
us_census.drop_duplicates (inplace = True)
```

```python
plt.scatter (us_census["Female"], us_census["Income"])
plt.xlabel ('Salary $', fontsize = 16)
plt.ylabel ('Female Population', fontsize = 16)
plt.show()
```
